```python
## (f) 
def Nheads(x, p=0.5, return_tails=False):
    h = 0
    t = 0
    for i in x:
        if i>p:
            h += 1
        else:
            t += 1
    if return_tails==True:
        return h, t
    else:
        return h
```

```python
# find counts and frequencies
counts = {} #counts
DF = {} #distribution frequencies

for i in N:
    counts[i] = []
    DF[i] = []
    counts[i], DF[i] = np.unique(k[i], return_counts=True)
    counts[i] = np.array(counts[i])
    DF[i] = np.array(DF[i])/i


#plot 
plt.figure(figsize=(15, 4))

for Nfig in range(3):
    index = 130+(Nfig+1)
    plt.subplot(index)
    plt.plot(counts[N[Nfig]], DF[N[Nfig]], '.-', color="purple", alpha=0.5, label='Empirical')
    plt.title('50 coin toss for {:.0f} runs'.format(N[Nfig]), size='large')
    plt.xlabel('# of heads (x)', size='large'); plt.ylabel('P(x)', size='large')
    plt.legend()
```

```python
#Firstly do not use built in functions!!
#functions to calculate mean and variance
def _mean_(x, px):
    return sum(x*px)

def _var_(x, px):
    mu = sum(x*px)
    return sum((x-mu)**2*px)

mean_coded, var_coded = {}, {} #coded
mean, var = {}, {} #built in

for i in N:
    mean_coded[i] = _mean_(counts[i], DF[i])
    var_coded[i] = _var_(counts[i], DF[i])
    mean[i] = np.mean(k[i])
    var[i] = np.var(k[i])

print(mean_coded, var_coded)
print(mean, var)
```

## 3 Random walks
### Mean Squared Displacement

```python
## _rw_(x, p): returns the cummulative step vector for the input random variable x, with bias probability
def _rw_(x, p):
    steps = np.zeros(len(x))
    for i in range(len(x)):
        if x[i]>(1-p):
            steps[i] = 1
        else:
            steps[i] = -1
    return np.cumsum(steps)
```

```python
## (a)
t0 = default_timer()
M = 50 ## coins tossed at each run
N = [10, 20 , 30, 40, 50, 100]  ## number of runs
p = 0.5


## (c)
Rmean = []
STD = []
VAR = []
MSD = []
XN = []

plt.figure(figsize=(6,2.5*len(N)))
ith = 0 ## index for subplot

for Nsteps in N:
    xN = []
    ith += 1
    for i in range(M):
        x = 0
        for n in range(Nsteps):
            r = np.random.rand()
            if r > p:
                x -= 1
            else:
                x += 1
        xN.append(x)
    
    ## Compute end-to-end and MSD
    xN = np.array(xN)
    Rmean.append(np.mean(xN))
    STD.append(np.std(xN))
    VAR.append(np.var(xN))
    MSD.append(np.mean(xN**2))

    ## plot
    plt.subplot(len(N), 1, ith)
    plt.hist(xN, density=True, label="N={:.0f}".format(Nsteps))
    plt.legend()
tf = default_timer()
print(f"time taken: {tf-t0:.4f} seconds")
```

```python
## _rw_(x, p): returns the cummulative step vector for the input random variable x, with bias probability
def _rw_(x, p):
    steps = np.zeros(len(x))
    for i in range(len(x)):
        if x[i]>(1-p):
            steps[i] = 1
        else:
            steps[i] = -1
    return np.cumsum(steps)
```

```python
## simulate M=1000 rw's for N=150 steps and bias probability of stepping to the right of p=0.5
t0 = default_timer()
M = 1000 ## #of rw's
N = 150 ## time step
p = 0.5

RW = np.zeros((M, N))

for row in range(M):
  r = np.random.rand(N)
  RW[row, :] = _rw_(r, p)
tf = default_timer()
print("time taken: ", tf-t0, "seconds")
```

```python
## plot trajectories
for i in range(M):
    plt.plot(RW[i,:])
    plt.ylabel("x")
    plt.xlabel("step (N)")
```

**End to end distance:**

$\langle (r(N)-r(N_0)) \rangle = \frac{1}{N}\sum_{N_i}^N (r(N_i)-r(N_0))$


**Mean-Squared-Displacement**:

$\langle (r(N)-r(N_0))^2 \rangle = \frac{1}{N}\sum_{N_i}^N (r(N_i)-r(N_0))^2$

```python
## compute the mean at every timestep
Rmean = np.zeros(N) ## zeros vecrtor of shape 1 row and N columns
MSD = np.zeros(N) ## zeros vecrtor of shape 1 row and N columns

## populate the vectors with R-end-to-end and MDS at every time step N
for Nstep in range(N):
    Rmean[Nstep] = np.mean(RW[:,Nstep])
    MSD[Nstep] = np.mean(RW[:,Nstep]**2)
```

```python
## plot all trajectories
for i in range(M):
    plt.plot(RW[i,:], color="lightblue", alpha=0.5)

## plot R-end-to-end & plot MSD
plt.plot(Rmean, 'r', label="mean")
plt.plot(MSD, 'k', label="MSD")
plt.ylabel("x")
plt.xlabel("step (N)")
plt.legend()
```

### Histograms and distributions

\begin{align*}
\langle x\rangle &= E[x] = \frac{1}{N_{rw}}\sum_i^{N_{rw}}x_i \\
STD[x] &= \sqrt{\frac{1}{N_{rw}}\sum_i^{N_{rw}}(x_i-\langle x\rangle)^2}
\end{align*}

```python
## calculate the standard deviation
STD = np.zeros(N) ## array of N columns & 1 row
STD2 = np.zeros(N)

for Nsteps in range(N):
    std = 0
    STD2[Nsteps] = np.std(RW[:,Nsteps])
    for row in range(M):
        std += (RW[row,Nsteps]-Rmean[Nsteps])**2/M
        STD[Nsteps] = np.sqrt(std)
```

### find diffusion from position distributions
\begin{align*}
P(\Delta x_i, t) &= \frac{1}{\sqrt{4\pi Dt}}e^{-\frac{\Delta x_i^2}{4Dt}} \\
P(x, \sigma^2) &= \frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x-\langle x\rangle)^2}{2\sigma^2}}
\end{align*}


From the equations above we find that the probability distribution for one random walker $x_i$ is a Gaussian distribution with mean $\langle  x\rangle=0$ and standard deviation $\sigma^2=2Dt$. The probability for all random walkers to be in position $\Delta x$ is the product of all random probablity distributions for all random walkers.

\begin{align*}
P(\vec{x}, t) = P(\Delta x_1, t) \times P(\Delta x_2, t) \times \ldots \times P(\Delta x_M,t) \\
P(\vec{x}, t) = (4\pi D t)^{-M/2}e^{-\frac{1}{4Dt}\sum_i^Mx_i^2}
\end{align*}

```python
def _gaussian_(x, xmean, sigma):
    return (1/np.sqrt(2*np.pi*sigma**2))*np.e**(-(x-xmean)**2/(2*sigma**2))
```

Linear relantionship between $\sigma^2$ and $t$
\begin{align*}
\sigma^2(t) &= 2Dt
\end{align*}

For a single data point $t_i$ then we can find $D$ 
\begin{align*}
\sigma^2(t_i) &= 2Dt_i \\
\rightarrow D &= \frac{\sigma^2(t_i)}{2t_i}
\end{align*}

If we have all time-steps information then $D$ can be found using a linear fit
\begin{align*}
\sigma^2(t) &= 2Dt \\
y(x) &= mx+b
\end{align*}

where $\quad y(x)=\sigma^2(t), \quad b=0, \quad m=2D$, so the diffusion term is 

\begin{align*}
\boxed{D = \frac{m}{2}}
\end{align*}
